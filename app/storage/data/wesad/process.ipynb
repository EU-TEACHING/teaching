```python
import pickle
import pandas as pd
```

```python
s3 = pickle.load(open('app/storage/data/wesad/S2.pkl', 'rb'), encoding='latin1')
```

```python
len(s3['label'])
```

```python
import numpy as np
X = np.concatenate(
            [
        s3["signal"]["chest"]["ACC"],
        s3["signal"]["chest"]["Resp"],
        s3["signal"]["chest"]["EDA"],
        s3["signal"]["chest"]["ECG"],
        s3["signal"]["chest"]["EMG"],
        s3["signal"]["chest"]["Temp"],
    ],
    axis=1,
)
Y = s3["label"]
X = X[(Y > 0) & (Y < 5)]
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
Y = Y[(Y > 0) & (Y < 5)]
Y = Y - 1
Y[Y > 1] = 0
for i, y in enumerate(Y):
    if y == 1:
        print("index is", i)
        break

# create dataframe from X and Y
df = pd.DataFrame(X)#[i-5000:i+5000])

# normalize columns
#df = (df - df.mean()) / df.std()
df.columns = ["acc_x", "acc_y", "acc_z", "ecg", "emg",  "eda", "temp", "resp"]
df["label"] = Y#[i-5000:i+5000]
```

```python
# save dataframe
df.to_csv("app/storage/data/wesad/s2.csv", index=False)
```
